```python
import pandas as pd

# Mengganti missing values HCT dengan nilai median
median_HCT = data['HCT'].median()
data['HCT'].fillna(median_HCT, inplace=True)

# Mengganti missing values MCHC dengan nilai median
median_MCHC = data['MCHC'].median()
data['MCHC'].fillna(median_MCHC, inplace=True)
```

```python
import pandas as pd

# Fungsi untuk mengganti missing value berdasarkan kondisi MCHC
def replace_rbc_hgb_conc(mchc_value):
    if mchc_value is None:
        return None  # Biarkan missing value pada 'RBC_Hgb_Conc' jika MCHC juga missing
    elif mchc_value < 31:
        return 'Low MCHC'
    elif 31 <= mchc_value <= 35:
        return 'Normal MCHC'
    else:
        return 'High MCHC'

# Mengganti missing value dalam kolom 'RBC_Hgb_Conc' dengan fungsi yang ditentukan
data['RBC_Hgb_Conc'] = data['MCHC'].apply(replace_rbc_hgb_conc)

# Menampilkan DataFrame setelah mengganti missing value
print("DataFrame setelah mengganti missing value:")
data

# Menampilkan baris-baris yang diminta
indices = [3084] + list(range(3661, 3674)) + list(range(4574, 4581)) + [4707]
selected_rows = data.iloc[indices]
selected_rows
```

# One Hot Encoding

```python
# Import Library yang akan digunakan
import numpy as np

max_k = 13 # Jumlah kategori paling banyak pada variabel yang akan di encode

# Daftar kolom yang akan dilakukan one-hot encoding
columns_to_encode = ['Age_months_pentachotomized']

# Lakukan one-hot encoding untuk setiap kolom
for col in columns_to_encode:
    top_values = data2[col].value_counts().nlargest(max_k).index
    for value in top_values:
        data2[col + '_' + str(value)] = (data2[col] == value).astype(int)

# Hapus kolom asli yang telah di-encode
data_encoded = data2.drop(columns=columns_to_encode)
data_encoded
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Inisialisasi model Random Forest Classifier
model3 = RandomForestClassifier()

# Melakukan fitting pada data training
model3.fit(X_train, y_train)

# Melakukan prediksi pada data training
y_train_pred = model3.predict(X_train)

# Melakukan prediksi pada data test
y_test_pred = model3.predict(X_test)

# Menghitung akurasi pada data training
train_accuracy = accuracy_score(y_train, y_train_pred)
print("Akurasi pada data training: ", train_accuracy)

# Menghitung akurasi pada data test
test_accuracy = accuracy_score(y_test, y_test_pred)
print("Akurasi pada data test: ", test_accuracy)
```

## K-Means with PCA

```python
import pandas as pd

# Membuat matriks korelasi
korelasi_matrix = dkluster.corr()

# Menampilkan matriks korelasi
korelasi_matrix
```

```python
korpca = np.array(korelasi_matrix)
print(korpca)
```

```python
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd

# Inisialisasi PCA dengan jumlah komponen utama yang diinginkan
pca = PCA(n_components=3)
pca.fit(korpca)

# Eigenvalues
eigenvalues = pca.explained_variance_

# Variance Proportion
variance_proportion = pca.explained_variance_ratio_

# Membulatkan variance proportion menjadi 2 angka di belakang koma
variance_proportion = np.round(variance_proportion, 2)

# Cumulative Variance Proportion
cumulative_variance_proportion = np.cumsum(variance_proportion)

# Membulatkan cumulative variance proportion menjadi 2 angka di belakang koma
cumulative_variance_proportion = np.round(cumulative_variance_proportion, 2)

# Menyimpan hasil dalam DataFrame
result_df = pd.DataFrame({
    'Eigenvalues': eigenvalues,
    'Variance Proportion': variance_proportion,
    'Cumulative Variance Proportion': cumulative_variance_proportion
}, index=['PC1', 'PC2', 'PC3'])

print(result_df)
```
